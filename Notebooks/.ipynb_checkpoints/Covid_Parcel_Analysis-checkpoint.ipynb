### Data Cleaning & Feature Engineering

```python

df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Week'].astype(str) + '1', format='%G%V%u')
df.head()

# creates a new column named Date which contains the actual calendar date for the start of the parcel shipping week.
```

```python

# Step 1: Total volume per customer for 2019
cust_vol_2019 = df[df['Year'] == 2019].groupby('CustomerID')['Parcels'].sum().reset_index()
cust_vol_2019.rename(columns={'Parcels': 'Cust_2019_Volume'}, inplace=True)

# Step 2: Total volume per customer for 2020
cust_vol_2020 = df[df['Year'] == 2020].groupby('CustomerID')['Parcels'].sum().reset_index()
cust_vol_2020.rename(columns={'Parcels': 'Cust_2020_Volume'}, inplace=True)

# Step 3: Define grouping logic
def assign_group(volume):
    if volume > 500_000:
        return 'Enterprise'
    elif 200_000 <= volume <= 500_000:
        return 'Large'
    elif 10_000 <= volume < 200_000:
        return 'Medium'
    elif 1_000 <= volume < 10_000:
        return 'Small'
    else:
        return 'Other'
        
# Step 4: Assign group based on 2019 and 2020 volumes
cust_vol_2019['Customer_Group_2019'] = cust_vol_2019['Cust_2019_Volume'].apply(assign_group)
cust_vol_2020['Customer_Group_2020'] = cust_vol_2020['Cust_2020_Volume'].apply(assign_group)

# Step 5: Merge back into the main DataFrame
df = df.merge(cust_vol_2019[['CustomerID', 'Customer_Group_2019']], on='CustomerID', how='left')
df = df.merge(cust_vol_2020[['CustomerID', 'Customer_Group_2020']], on='CustomerID', how='left')
df.head()


# This calculates the total annual parcel volume for each customer in 2019 and 2020, 
# assigns a corresponding Customer Size Group (Enterprise, Large, Medium, Small, or Other) 
# then based on those annual volumes, and merges the resulting group labels back into the main DataFrame.
```

```python
# Step 1: Define the discount structure
discount_map = {
    'Enterprise': 0.22,
    'Large': 0.17,
    'Medium': 0.10,
    'Small': 0.04,
    'Other': 0.00
}

# Step 2: Assign discount based on 2019 and 2020 customer groups
df['Discount_2019'] = df['Customer_Group_2019'].map(discount_map)
df['Discount_2020'] = df['Customer_Group_2020'].map(discount_map)
df.head()
```

## Overall Business Performance: 2019 vs. 2020

Calculate the change in **Volume and Customer base** between week 12-53 2019 and 2020

```python
# Filter for Week 12â€“53 of 2019 (COVID Observation Period Equivalent)
df_2019_sp = df[(df['Year'] == 2019) & (df['Week'] >= 12)]

# Filter for Week 12â€“53 of 2020 (COVID Observation Period)
df_2020_covid = df[(df['Year'] == 2020) & (df['Week'] >= 12)]

# Aggregate volume for both periods
volume_2019 = df_2019_sp['Parcels'].sum()
volume_2020 = df_2020_covid['Parcels'].sum()

# Calculate the Year-over-Year (YoY) volume change
volume_change = ((volume_2020 - volume_2019) / volume_2019) * 100

print(f"Total Parcels (Week 12-53, 2019): {volume_2019:,.0f}")
print(f"Total Parcels (Week 12-53, 2020): {volume_2020:,.0f}")
print(f"\nOverall Volume Change (YoY COVID Period): +{volume_change:.2f}%")
```

```python
# customers base difference between the two periods

customers_2019 = df_2019_sp['CustomerID'].nunique()
customers_2020 = df_2020_covid['CustomerID'].nunique()
customer_change = ((customers_2020 - customers_2019) / customers_2019) * 100


print(f"Customer Total (Week 12-53, 2019: {customers_2019:,.0f}")
print(f"Customer Total (Week 12-53, 2020): {customers_2020:,.0f}")
print(f"\nOverall Change (YoY COVID Period): {customer_change:.2f}%")
```

```python
# Step 2: Create the summary DataFrame (without revenue)
summary = pd.DataFrame({
    'Metric': ['Parcel Volume', 'Active Customers'],
    '2019 (Wk12â€“53)': [volume_2019, customers_2019],
    '2020 (Wk12â€“53)': [volume_2020, customers_2020],
    '% Change': [f"{volume_change:.2f}%", f"{customer_change:.2f}%"]
})

# Step 3: Format large numbers for readability
summary['2019 (Wk12â€“53)'] = summary['2019 (Wk12â€“53)'].apply(lambda x: f"{int(round(x)):,}")
summary['2020 (Wk12â€“53)'] = summary['2020 (Wk12â€“53)'].apply(lambda x: f"{int(round(x)):,}")

# Step 4: Display the final summary table
summary
```

## 	Q3 What events within the COVID timeline may have contributed to the change?

```python
# Key COVID timeline events with approximate week numbers
covid_events = {
    5: 'Jan 27: 1st Case',
    11: 'Mar 11: Pandemic Declared',
    12: 'Mar 16-18: WFH & Border Restriction',
    13: 'Mar 23: Post_Covid Delivery Standard',
    15: 'Apr 6: C.E.R.B Launched'
}
```

```python
plt.figure(figsize=(14,6))

# Plot weekly parcel volumes for 2018, 2019, 2020
for year in [2019, 2020]:
    if year in weekly_volume_pivot.columns:
        plt.plot(weekly_volume_pivot.index, weekly_volume_pivot[year], label=str(year), linewidth=2)

# Highlight COVID events with vertical lines
for week, label in covid_events.items():
    plt.axvline(x=week, color='gray', linestyle='--', linewidth=1)
    plt.text(week + 0.2, plt.ylim()[1]*0.95, label, rotation=90, fontsize=9, color='black')

# Highlight initial divergence
plt.axvline(x=11, color='red', linestyle='--', linewidth=2, label='Week 11 - Initial Impact')

# Labels and legend
plt.title('Weekly Parcel Volume with COVID Events (2018â€“2020)', fontsize=14, fontweight='bold')
plt.xlabel('Week Number')
plt.ylabel('Total Parcels')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Timeline Analysis: Events Driving Parcel Volume Change

The significant and sustained upward divergence in 2020 parcel volume, starting abruptly in **Week 11**, was a direct result of key global and domestic events that fundamentally altered consumer and business behavior.

* **Week 11 (Mar 11): Pandemic Declared:** The official declaration initiated a universal sense of crisis, triggering the first wave of bulk purchasing and stockpiling, which immediately caused the volume trend to accelerate.
* **Week 12 (Mar 16-18): WFH & Border Restriction:** The rapid implementation of Work From Home (WFH) mandates and international border restrictions forced the closure of physical retail. This massive shock **redirected consumer spending almost entirely to e-commerce**, translating directly into sustained, elevated parcel volumes. 
* **Week 13 (Mar 23): Post-COVID Delivery Standard:** The industry's quick implementation of new delivery standards indicates that the high-volume environment was immediately accepted as the new operational normal in the post-shock period.
* **Week 15 (Apr 6): C.E.R.B Launched:** The introduction of major government income support programs injected significant financial liquidity into the consumer market. This support was critical in enabling and sustaining the high rate of online spending throughout the rest of the year, preventing a volume crash after the initial panic-buying phase.

**Conclusion:** The volume change was driven by a rapid, cascading series of external shocks that accelerated the existing shift to e-commerce, turning a traditional parcel delivery business into a key component of the new pandemic economy.

## Q6 What percent of each customer group is growing, moderately growing, and declining during the COVID observation period? Â Â 

```python
df.head()
```

**Define ISGR and filter periods**

```python
ISGR = 11.40  # Industry Standard Growth Rate from pre-COVID

# Pre-COVID: Week 12, 2019 â†’ Week 11, 2020
df_precovid = df[
    ((df['Year'] == 2019) & (df['Week'] >= 12)) |
    ((df['Year'] == 2020) & (df['Week'] <= 11))
]

# COVID Observation: Week 12, 2020 â†’ Week 53, 2020
df_covid = df[(df['Year'] == 2020) & (df['Week'] >= 12)]

# 2019 same period baseline for COVID comparison
df_2019_sp = df[(df['Year'] == 2019) & (df['Week'] >= 12)]
```

**Aggregate parcel volumes per customer**

```python
# Total parcels per customer in pre-COVID period
cust_precovid = df_precovid.groupby('CustomerID')['Parcels'].sum().reset_index()
cust_precovid.rename(columns={'Parcels': 'Parcels_PreCOVID'}, inplace=True)

# Total parcels per customer during COVID period (2020)
cust_2020_covid = df_covid.groupby('CustomerID')['Parcels'].sum().reset_index()
cust_2020_covid.rename(columns={'Parcels': 'Parcels_2020_COVID'}, inplace=True)

# Total parcels per customer for 2019 same period (baseline)
cust_2019_sp = df_2019_sp.groupby('CustomerID')['Parcels'].sum().reset_index()
cust_2019_sp.rename(columns={'Parcels': 'Parcels_2019_COVID'}, inplace=True)
```

**Merge all parcel totals**

```python
cust = cust_precovid.merge(cust_2019_sp, on='CustomerID', how='outer') \
                    .merge(cust_2020_covid, on='CustomerID', how='outer') \
                    .fillna(0)
```

**Merge customer group info**

```python
cust = cust.merge(
    df[['CustomerID','Customer_Group_2019']].drop_duplicates(),
    on='CustomerID',
    how='left'
)

cust = cust.merge(
    df[['CustomerID','Customer_Group_2020']].drop_duplicates(),
    on='CustomerID',
    how='left'
)
```

**Compute % growth during COVID period**

```python
cust['Growth_%'] = (
    (cust['Parcels_2020_COVID'] - cust['Parcels_2019_COVID']) /
    cust['Parcels_2019_COVID'].replace(0, 1) * 100
)
```

**Classify customers by growth**

```python
def classify_customer(row, ISGR=ISGR):
    pre = row['Parcels_PreCOVID']
    c2019 = row['Parcels_2019_COVID']
    c2020 = row['Parcels_2020_COVID']
    growth = row['Growth_%']
    grp_2019 = row['Customer_Group_2019']
    grp_2020 = row['Customer_Group_2020']

    group_order = ['Other', 'Small', 'Medium', 'Large', 'Enterprise']

    # New Customers
    if pre == 0 and c2020 > 0:
        return "New Customers"

    # Lost Customers
    if pre > 0 and c2020 == 0:
        return "Lost Customers"

    # Declining Customers: negative growth OR downgrade in group
    if pre > 0 and (growth < 0 or group_order.index(grp_2020) < group_order.index(grp_2019)):
        return "Declining Customers"

    # High Growth Customers
    if growth > ISGR:
        return "High Growth Customers"

    # Stable Customers
    if 0 <= growth <= ISGR:
        return "Stable Customers"

    return "Stable Customers"

cust['Growth_Category'] = cust.apply(classify_customer, axis=1)
```

**Pivot table by customer group**

```python
summary_counts = cust.pivot_table(
    index='Customer_Group_2020',
    columns='Growth_Category',
    values='CustomerID',
    aggfunc='count',
    fill_value=0
)

summary_pct = summary_counts.div(summary_counts.sum(axis=1), axis=0) * 100
summary_pct = summary_pct.round(2)

# Display
summary_pct
```

## ** Q7 What percent of each customer group are new customers during COVID?**

```python
# 1. Get the unique set of Customer IDs in 2019
customers_2019 = set(df[df['Year'] == 2019]['CustomerID'].unique())

# 2. Get the unique set of Customer IDs in 2020
customers_2020 = set(df[df['Year'] == 2020]['CustomerID'].unique())

# 3. Use set difference to find customers in 2020 who are NOT in 2019
# The difference() method returns items in the first set (2020) that are not found in the second set (2019).
new_customers_2020 = customers_2020.difference(customers_2019)

# 4. Convert the set back to a list/DataFrame for review or further analysis
print(f"Total Unique Customers in 2019: {len(customers_2019)}")
print(f"Total Unique Customers in 2020: {len(customers_2020)}")
print(f"\nNumber of New Customers in 2020: {len(new_customers_2020)}")
print("\n--- Sample of New 2020 Customer IDs ---")
# Displaying the first 10 new customer IDs
print(list(new_customers_2020)[:10])

# To create a DataFrame containing only the new customers' records:
df_new_customers = df[df['CustomerID'].isin(new_customers_2020)]
```

```python
# Step 1: Filter COVID period data
covid_period = df[(df['Year'] == 2020) & (df['Week'].between(12, 53))]

# Step 2: Get unique customers in COVID period
customers_2020_covid = set(covid_period['CustomerID'].unique())

# Step 3: Get unique customers in 2019 (baseline)
customers_2019 = set(df[df['Year'] == 2019]['CustomerID'].unique())

# Step 4: Identify new customers in COVID period
new_customers = customers_2020_covid - customers_2019

# Step 5: Create DataFrame for new customers
new_customers_df = covid_period[covid_period['CustomerID'].isin(new_customers)][['CustomerID','Customer_Group_2020']].drop_duplicates()

# Step 6: Count new customers per group
new_counts = new_customers_df.groupby('Customer_Group_2020')['CustomerID'].nunique().reset_index()
new_counts.rename(columns={'CustomerID':'New_Customers'}, inplace=True)

# Step 7: Calculate % of each group
total_customers_by_group = covid_period.groupby('Customer_Group_2020')['CustomerID'].nunique().reset_index()
total_customers_by_group.rename(columns={'CustomerID':'Total_Customers'}, inplace=True)

new_counts = new_counts.merge(total_customers_by_group, on='Customer_Group_2020')
new_counts['% of Group'] = (new_counts['New_Customers'] / new_counts['Total_Customers'] * 100).round(2)

new_counts[['Customer_Group_2020','New_Customers','% of Group']]
```

## ** Q8 What percent of 2019 customers in each group did we lose during the COVID observation period?**

```python
# Step 1: Total parcels per customer per year
cust_total_2019 = df[df['Year'] == 2019].groupby('CustomerID')['Parcels'].sum().reset_index()
cust_total_2020 = df[df['Year'] == 2020].groupby('CustomerID')['Parcels'].sum().reset_index()

# Step 2: Merge customer totals using outer join to keep all customers
cust_merge = pd.merge(cust_total_2019, cust_total_2020, on='CustomerID', how='outer', suffixes=('_2019','_2020')).fillna(0)

# Step 3: Identify lost customers (Parcels_2020 == 0 and Parcels_2019 > 0)
cust_merge['Lost'] = (cust_merge['Parcels_2019'] > 0) & (cust_merge['Parcels_2020'] == 0)

# Step 4: Identify new customers (Parcels_2019 == 0 and Parcels_2020 > 0)
cust_merge['New'] = (cust_merge['Parcels_2019'] == 0) & (cust_merge['Parcels_2020'] > 0)

# Step 5: Map customer groups for 2019 and 2020
cust_merge = cust_merge.merge(df[['CustomerID','Customer_Group_2019']].drop_duplicates(), on='CustomerID', how='left')
cust_merge = cust_merge.merge(df[['CustomerID','Customer_Group_2020']].drop_duplicates(), on='CustomerID', how='left')

# Step 6: Summarize Lost customers per 2019 group
lost_summary = cust_merge.groupby('Customer_Group_2019')['Lost'].sum().reset_index()
lost_summary.rename(columns={'Lost':'Lost_Customers'}, inplace=True)
lost_summary['% of 2019 Group'] = (lost_summary['Lost_Customers'] / 
                                   cust_merge.groupby('Customer_Group_2019')['CustomerID'].nunique().values * 100).round(2)

# Step 7: Summarize New customers per 2020 group
new_summary = cust_merge.groupby('Customer_Group_2020')['New'].sum().reset_index()
new_summary.rename(columns={'New':'New_Customers'}, inplace=True)
new_summary['% of 2020 Group'] = (new_summary['New_Customers'] / 
                                  cust_merge.groupby('Customer_Group_2020')['CustomerID'].nunique().values * 100).round(2)

lost_summary
```

## What was the overall impact of COVID on volumes and revenue by customer group?

```python
# Step 1: Filter dataset for Week 11â€“53 (same period)
df_2019_sp = df[(df['Year'] == 2019) & (df['Week'].between(12, 53))]
df_2020_covid = df[(df['Year'] == 2020) & (df['Week'].between(12, 53))]

# Step 2: Aggregate total parcels and revenue per customer group
summary_2019 = df_2019_sp.groupby('Customer_Group_2020').agg(
    Volume_2019=('Parcels','sum'),
    Revenue_2019=('Parcels', lambda x: sum(x * 22 * (1 - df_2019_sp.loc[x.index, 'Discount_2019'])))
).reset_index()

summary_2020 = df_2020_covid.groupby('Customer_Group_2020').agg(
    Volume_2020=('Parcels','sum'),
    Revenue_2020=('Parcels', lambda x: sum(x * 22 * (1 - df_2020_covid.loc[x.index, 'Discount_2020'])))
).reset_index()

# Step 3: Merge 2019 and 2020 summaries
df_same_period = pd.merge(summary_2019, summary_2020, on='Customer_Group_2020', how='outer').fillna(0)

# Step 4: Calculate % change
df_same_period['Volume_%Change'] = ((df_same_period['Volume_2020'] - df_same_period['Volume_2019']) /
                                    df_same_period['Volume_2019'] * 100).round(2)
df_same_period['Revenue_%Change'] = ((df_same_period['Revenue_2020'] - df_same_period['Revenue_2019']) /
                                     df_same_period['Revenue_2019'] * 100).round(2)

# Display the final dataframe
df_same_period
```

```python
# Set the width of the bars
bar_width = 0.35
x = np.arange(len(df_same_period['Customer_Group_2020']))

# Plot Volume comparison
plt.figure(figsize=(12,6))
plt.bar(x - bar_width/2, df_same_period['Volume_2019'], width=bar_width, label='Volume 2019', alpha=0.8)
plt.bar(x + bar_width/2, df_same_period['Volume_2020'], width=bar_width, label='Volume 2020', alpha=0.8)
plt.xticks(x, df_same_period['Customer_Group_2020'])
plt.xlabel('Customer Group')
plt.ylabel('Parcel Volume')
plt.title('Parcel Volume by Customer Group: 2019 vs 2020 (Week 12â€“53)')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Plot Revenue comparison
plt.figure(figsize=(12,6))
plt.bar(x - bar_width/2, df_same_period['Revenue_2019'], width=bar_width, label='Revenue 2019', alpha=0.8)
plt.bar(x + bar_width/2, df_same_period['Revenue_2020'], width=bar_width, label='Revenue 2020', alpha=0.8)
plt.xticks(x, df_same_period['Customer_Group_2020'])
plt.xlabel('Customer Group')
plt.ylabel('Revenue ($)')
plt.title('Revenue by Customer Group: 2019 vs 2020 (Week 12â€“53)')  # <-- fixed closing quote
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
