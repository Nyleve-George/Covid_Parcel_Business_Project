# 3. Overall Volume Trends

```python
weekly = (
    df[df['Year'].isin([2018, 2019, 2020, 2021])]
      .groupby(['Year', 'Week'], as_index=False)['Parcels']
      .sum()
)
```

```python
pivot = weekly.pivot(index='Week', columns='Year', values='Parcels')
```

```python
plt.figure(figsize=(12,5))
plt.plot(pivot.index, pivot[2018], label="2018")
plt.plot(pivot.index, pivot[2019], label="2019")
plt.plot(pivot.index, pivot[2020], label="2020")
plt.plot(pivot.index, pivot[2021], label="2021")
plt.axvline(11, color="grey", linestyle="--")  # pre-COVID boundary
plt.legend()
plt.title("Weekly Parcel Volume: 2019 vs 2020")
plt.xlabel("Week")
plt.ylabel("Total Parcels")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(pivot.index, pivot[2019], label="2019")
plt.plot(pivot.index, pivot[2020], label="2020")
plt.axvline(11, color="grey", linestyle="--")  # pre-COVID boundary
plt.legend()
plt.title("Weekly Parcel Volume: 2019 vs 2020")
plt.xlabel("Week")
plt.ylabel("Total Parcels")
plt.show()
```
